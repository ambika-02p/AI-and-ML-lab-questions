```python
#counting the number of students in each semester.
print(df["semester"].value_counts())
```

```python
#finding outliers in marks column using both iqr and z-score.
q1 = np.percentile([90, 100], 25)
q3 = np.percentile( [90, 100], 75)
iqr = q3 - q1
lower = [ (-1.5)*iqr] + q1
upper = [ (1.5)*iqr ] + q3
print(lower)
print(upper)
```

```python
#correlation between marks and study hours.
marks_studyhours = df["marks"].corr(df["studyhours"])
print("correlation is:", marks_studyhours)
```
