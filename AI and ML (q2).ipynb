```python
#printing the dataset
import pandas as pd 
import numpy as np

data1 = pd.read_csv("22yztw.csv")
print(data1)
```

```python
data2 = pd.read_csv("6svfo1.csv")
print(data2)
```

```python
#merging the two datasets.
merged_data = pd.merge(data1, data2, on = "patient_id")
print(merged_data)
```

```python
merged_data.to_csv("merged_data.csv", index=False)
```

```python
merged_data['gender'] = merged_data['gender'].replace(['F','M'] , ['Female','Male'])
merged_data['height_cm'] = merged_data['height_cm'].replace('abc', np.nan)
merged_data["age"] = merged_data["age"].replace(-5, 5)
merged_data["name"] = merged_data["name"].fillna("unknown")
merged_data["bp"] = merged_data["bp"].replace("high" , "140/90")
merged_data["admission_date"] = merged_data["admission_date"].replace("2024-01-15", " 15/01/2024")
merged_data["admission_date"] = merged_data["admission_date"].replace("2024/01/16", "16/01/2024")
merged_data = merged_data[['record_id', 'patient_id', 'admission_date', 'name', 'age', 'gender', 'height_cm', 'city', 'bp', 'sugar_level', 'diagnosis', 'visit_cost']]

print(merged_data)
```

```python
merged_data.to_csv("merged_data.csv", index=False)
```
