```python
def dfs (start, goal, graph, vis = None):
    if vis is None:
        vis = set()

    print(start)

    if start == goal:
        return True

    vis.add(start)
    for child in graph [start]:
        if child not in vis:
            if dfs( child, goal, graph, vis):
                return True

    return False

graph = {}
n = int(input("Enter the number of nodes:"))
for i in range (n):
    node = input("Enter the name of nodes:")
    graph[node] = []
    n = int(input(f"Enter the number of neighbours of {node}:"))
    for j in range (n):
        temp = input()
        graph[node].append(temp)

start_node = input("enter the starting node:")
end_node = input("enter the ending node:")

dfs(start_node, end_node, graph)
```
